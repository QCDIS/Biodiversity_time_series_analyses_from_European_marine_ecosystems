# Biodiversity time series analyses from European marine ecosystems

![images/biotisan_euromarec_flowchart.png](images/biotisan_euromarec_flowchart.png)

To run the workflow do the following:  

#### Prepare the data and access key on MinIO
This virtual lab uses MinIO to upload the data that will be analyzed and store the results of the workflow. You can log into MinIO at https://scruffy.lab.uvalight.net:9001 using the same credentials as you're using to log into NaaVRE. In order to store and retrieve data from MinIO, you will need an access key. Create an access key [in MinIO](https://scruffy.lab.uvalight.net:9001/access-keys). Make sure to store the access key and secret key. 

You can run the workflow with sample data or with your own data. In case you want to use your own data, in MinIO, go to Object Browser, naa-vre-user-data, open the directory named after the e-mail address you've used to log into NaaVRE. Upload your own data, using the _Excel_ template.  

#### Run the workflow in NaaVRE
In the file browser of NaaVRE, go to "Virtual Labs/Biodiversity Time Series Analyses/Git public/Biodiversity_time_series_analyses_from_European_marine_ecosystems/workflows".

![images/open_workflow.png](images/open_workflow.png)

Open _"Biodiversity_time_series_analyses.naavrewf"_

Press "Run"

![images/run_workflow.png](images/run_workflow.png)

Press "Use default parameter values"

![images/default_parameter_values.png](images/default_parameter_values.png)

Set _param_use_dummy_data_ to "1" if you want to use the dummy data. Set it to 0 if you'fe uploaded 

![images/set_param_use_dummy_data.png](images/set_param_use_dummy_data.png)

Fill in the MinIO access key and secret key you've just generated.

Press "Run"

![images/press_run.png](images/press_run.png)

Follow the link to the workflow execution view in Argo workflow, or open your workflow here: https://staging.demo.naavre.net/argowf/workflows by finding the workflow that has your e-mail address.

![images/track_workflow_here.png](images/track_workflow_here.png)

Once the workflow has succeeded, the results from the workflow are stored in MinIO. 

#### Retrieve the output data from MinIO
In MinIO, go to Object Browser, naa-vre-user-data, open the directory named after the e-mail address you've used to log into NaaVRE.

Here you should see a results file that has the name picked by the param.

![images/results_in_minio_user_data.png](images/results_in_minio_user_data.png)
